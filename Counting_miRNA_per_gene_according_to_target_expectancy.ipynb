```python
import pandas as pd

# ---- load data ----
# if tab-separated
df = pd.read_csv("/content/at mi gene.csv")

# if comma-separated instead, use:
# df = pd.read_csv("mirna_targets.csv")

# clean spaces
df["Gene"] = df["Gene"].astype(str).str.strip()
df["Expectation"] = pd.to_numeric(df["Expectation"], errors="coerce")

# ---- classify probability ----
high_scores = [3, 4]
low_scores = [4.5, 5]

df["Probability"] = df["Expectation"].apply(
    lambda x: "High" if x in high_scores else ("Low" if x in low_scores else "Other")
)

# ---- counts ----
overall_counts = df["Probability"].value_counts()
print("Overall counts:")
print(overall_counts)

# ---- counts per miRNA ----
mirna_counts = df.groupby(["miRNA", "Probability"]).size().unstack(fill_value=0)
print("\nCounts per miRNA:")
print(mirna_counts)

# ---- counts per Gene ----
gene_counts = df.groupby(["Gene", "Probability"]).size().unstack(fill_value=0)
print("\nCounts per Gene:")
print(gene_counts)

# ---- save results ----
mirna_counts.to_csv("mirna_probability_counts.csv")
gene_counts.to_csv("gene_probability_counts.csv")
```
